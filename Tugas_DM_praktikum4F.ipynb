```python
#Import library pandas
import pandas as pd
#Mounting Drive
from google.colab import drive
drive.mount('/content/gdrive')

#Membaca file CSV dari Google Drive, melewati baris yang bermasalah
df = pd.read_csv("gdrive/My Drive/ColabDataMining/DataSetA.csv", on_bad_lines='skip')
df    #Menampilkan DataFrame yang telah diimpor
```

```python
# Mengambil semua nilai dalam DataFrame df sebagai array NumPy
df.values
```

```python
# Convert each element within the nested lists to string
data = [[str(element) for element in row] for row in df.values]
data
```

```python
#Import TransactionEncoder untuk mengubah data transaksi menjadi bentuk one-hot encoding
from mlxtend.preprocessing import TransactionEncoder

#Inisialisasi encoder transaksi
te = TransactionEncoder()
#Melakukan 'fit' untuk mempelajari item unik dan 'transform' untuk mengubah data transaksi menjadi one-hot encoding
te_data = te.fit(data).transform(data)
#Mengubah data one-hot encoding menjadi DataFrame dengan nama kolom sesuai item
df = pd.DataFrame(te_data, columns=te.columns_)
#Mengonversi nilai ke tipe integer untuk memudahkan analisis
df = df.astype(int)
#Menampilkan DataFrame hasil one-hot encoding
df
```

```python
#Import fungsi apriori dan association_rules dari mlxtend untuk analisis asosiasi
from mlxtend.frequent_patterns import apriori, association_rules

#Menemukan itemset yang sering muncul berdasarkan min_support yang sangat rendah
frequent_itemsets = apriori(df, min_support=0.00001, use_colnames=True)

#Menghasilkan aturan asosiasi berdasarkan itemset dengan lift minimal 0.1
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.1)
```
